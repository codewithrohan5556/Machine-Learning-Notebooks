```python
import os
import glob
import warnings
csv_files = glob.glob(os.path.join(path, '*.csv'))

if csv_files:
    actual_csv_path = csv_files[0]
    df = pd.read_csv(actual_csv_path, low_memory=False)
else:
    print(f"Error: No CSV files found in the directory: {path}")
```

```python
df = df.dropna().reset_index(drop=True)
```

```python
from datetime import datetime
df['Formatted Date'] = pd.to_datetime(df['Formatted Date'], utc=True)
```

```python
df.drop("Formatted Date",axis=1,inplace=True)
df.drop("Daily Summary",axis=1,inplace=True)
df.drop("Summary",axis=1,inplace=True)
```

```python
df['Precip Type'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
le.fit(df['Precip Type'])
```

```python
df['Precip Type'] = le.transform(df['Precip Type'])
df.head()
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
X = df.drop("Temperature (C)", axis=1)
y = df["Temperature (C)"]
```

```python
X.shape
```

```python
y.shape
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
print("MAE",mean_absolute_error(y_test,y_pred))
print("MSE",mean_squared_error(y_test,y_pred))
print("R2 Score",r2_score(y_test,y_pred))
```

```python
import matplotlib.pyplot as plt
import numpy as np


features = X.columns
target = 'Temperature (C)'

fig, axes = plt.subplots(3, 3, figsize=(15, 15))
axes = axes.ravel()


for idx, feature in enumerate(features):
    axes[idx].scatter(df[feature], df[target], alpha=0.3, s=5)
    axes[idx].set_xlabel(feature, fontsize=10)
    axes[idx].set_ylabel(target, fontsize=10)
    axes[idx].set_title(f'{feature} vs {target}', fontsize=12)
    axes[idx].grid(True, alpha=0.3)


if len(features) < len(axes):
    axes[len(features)].set_visible(False)

plt.tight_layout()
plt.show()
```
